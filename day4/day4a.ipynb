### Class
---

```python
class Student():
    def __init__(self , name , usn , dept , clg , cgpa):
        self.name = name
        self.dept = dept
        self.usn = usn
        self.clg = clg
        self.cgpa = cgpa
    def __str__(self):
        return self.name + " " +  self.usn
    def intro(self):
        return "Name : " + self.name + "\nDept : " + self.dept + "\nUSN : " + self.usn  + "\nCollege : " + self.clg + "\nCGPA : " + str(self.cgpa)
     
```

#### object 
---

```python
std1 = Student("Gold D Roger", "1VI20CS096", "CSE", "VIT" , 10)
std2 = Student("Monkey .D Luffy ", "1VI20CS069", "PPP", "One Piece" , 6.9)
std3 = Student("Roronoa Zoro", "1VI20CS120", "SWRD", "UNKN" , 1.0)

print(std1 , std2 , std3)
print("\n")
print(std1.intro())
```
